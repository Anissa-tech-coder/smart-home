```python
import sys
sys.path.append('..')  # Pour accéder aux fichiers du parent

from config import ROOMS, COMFORT_LEVELS
from user_preferences import UserPreferencesManager
from actions import ActionGenerator
from energy_calculator import EnergyCalculator
from recommendation_engine import RecommendationEngine
import json

print("✅ Tous les imports réussis!")
```

```python
# Initialiser les composants
manager = UserPreferencesManager("../data/user_preferences.json")
generator = ActionGenerator()
calculator = EnergyCalculator(electricity_cost_per_kwh=0.15)
engine = RecommendationEngine(manager, generator, calculator)

print("✅ Composants initialisés!")
```

```python
# Créer l'utilisateur Tasnim
manager.create_user_profile("tasnim_001", "Tasnim")
manager.update_global_comfort_level("tasnim_001", "normal")

# État actuel simulé (capteurs IoT)
current_state = {
    "salon": {"temperature": 22, "brightness": 90},
    "chambre_1": {"temperature": 21, "brightness": 75},
    "chambre_2": {"temperature": 20, "brightness": 80},
    "cuisine": {"temperature": 22, "brightness": 100},
    "salle_bain": {"temperature": 22, "brightness": 85}
}

# Générer les recommandations
recommendations = engine.generate_recommendations(
    user_id="tasnim_001",
    current_state=current_state,
    max_recommendations=5
)

# Afficher le rapport
report = engine.generate_recommendation_report(recommendations)
print(report)
```

```python
print("\n" + "="*60)
print("DÉTAILS EN JSON")
print("="*60)
print(json.dumps(recommendations, indent=2, ensure_ascii=False))
```

```python
# Projections d'économies
daily = calculator.project_savings_per_day(generator.generate_all_recommendations(
    manager.get_user_preferences("tasnim_001"), 
    current_state
))

monthly = calculator.project_savings_per_month(generator.generate_all_recommendations(
    manager.get_user_preferences("tasnim_001"), 
    current_state
))

print("\n📊 ÉCONOMIES PRÉVUES:\n")
print(f"💡 Par JOUR (24h):")
print(f"   ⚡ {daily['total_energy_saved_kwh']} kWh")
print(f"   💰 {daily['total_cost_saved_euros']}€\n")

print(f"💡 Par MOIS (30 jours):")
print(f"   ⚡ {monthly['total_energy_saved_kwh']} kWh")
print(f"   💰 {monthly['total_cost_saved_euros']}€")
```
